```python
columns=['Occupation','Relationship','Race','Sex','Education','Workclass','Martial-Status']
```

```python
start=time.time()
for column in columns:
    df[column].value_counts()
    index=df[column].value_counts().index.tolist()
    values=df[column].value_counts().tolist()
    plt.barh(np.arange(len(index)),values,tick_label=index)
    plt.title(column)
    plt.show()
end=time.time()
print(end-start)
```
